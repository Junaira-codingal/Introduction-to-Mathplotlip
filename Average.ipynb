```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(x , y, marker = "*", mec = "r",ms=15, linestyle = "dashed",color="purple",linewidth=3)
plt.show()
```

```python
students_names=["snigdha","erina","maowa","nazia","tasmiya","labonno","esma","nishtha"]

students_marks=[35,50,20,45,25,40,25,40]
```

```python
marks_perc = []

for x in students_marks:
    res = (x/50)*100
    marks_perc.append(res)
    
marks_perc
```

```python
def line_chart_of_students_and_marks():
    plt.plot(students_names, students_marks, marker = "*", mec = "k",ms=15, linestyle = "dashed",color="purple",linewidth=3)
    plt.title("Students Marks Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Marks")
    plt.show()
    
line_chart_of_students_and_marks()
```

```python
def percentage_bar_chart():
    plt.bar(students_names, marks_perc,color="purple")
    plt.title("Students Percentage Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Percentage")
    plt.show()
    
percentage_bar_chart()
```
